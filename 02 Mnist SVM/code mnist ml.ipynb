```python
import numpy as np #thư viện numpy
import pandas as pd #thư viện đọc file csv
from sklearn import svm,metrics #thư viện ML trong đó SVM là một giải thuật và metrics để đánh giá mô hình phân loại
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score 
from sklearn.model_selection import train_test_split
```

```python
train = pd.read_csv('/content/sample_data/mnist_train_small.csv')# đọc file csv
train.head(8) # in ra 8 giá trị đầu tiên
```

```python
y_label = train['6'] # lấy cột label có nhãn 6
X_data = train.drop('6', axis = 1) #lấy các cột trừ cột label(6)
X_train_data = np.array(X_data)/255 # chia nhỏ tập dữ liệu dựa trên r,g,b[0,255]
print(X_train_data )
```

```python
X_train,X_test,y_train,y_test = train_test_split(X_train_data,y_label,test_size=0.2)
# tập train sẽ lấy 80% cho tập train và 20% cho tập test.
```

```python
# Code dùng bằng SVM
clf = svm.SVC(kernel='linear',C=100) # sử dụng tt SVM.SVC kiểu hồi quy tuyến tính với tham số phạt trong bài toán soft-mirgin là C 
clf.fit(X_train,y_train)# phân lớp tập (X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
#print(y_pred)#in dự đoán
print()# so sánh độ chính xác
print('accuracy score = ', metrics.accuracy_score(y_test,y_pred))
print(y_pred)#in dự đoán
```
