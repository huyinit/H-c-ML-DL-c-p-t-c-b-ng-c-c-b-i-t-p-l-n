```python
from sklearn.metrics import confusion_matrix # 
from sklearn.metrics import classification_report #báo cáo
# Tách dữ liệu thành các biến phụ thuộc và độc lập 
X = dataset.iloc[:, :-1].values 
y = dataset.iloc[:, -1].values

# Tách tập dữ liệu thành tập huấn luyện và tập kiểm tra 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0) # random_state tập test ngẫu nhiên

# Code dùng bằng SVM
#Support Vector Machine (SVM) là một thuật toán thuộc nhóm Supervised Learning (Học có giám sát)
from sklearn.svm import SVC

classifier = SVC(kernel='linear', probability=True) #Xác suất: Khi tạo lớp SVC để tính toán các ước tính xác suất bằng cách thiết lập probability=True:
classifier.fit(X_train, y_train) # phân lớp tập (X_train, y_train)
w = classifier.coef_ #hệ số
print('w = ', w)


# chú thích code :
#iloc :chọn đồng thời cả hai hàng và cột
# random_state : khởi tạo cho nó một cái mốc
```

```python
y_pred = classifier.predict(X_test) # dự đoán y_pred từ tập input X_test
print(y_pred)#in dự đoán
print()
```

```python
# so sánh độ chính xác
from sklearn.metrics import accuracy_score
print('accuracy is',accuracy_score(y_pred,y_test))
```

```python
# Tóm tắt các dự đoán được thực hiện bởi trình phân loại 
print(classification_report(y_test, y_pred))
#Macro-average precision là trung bình cộng của các precision theo class
#weighted avg :trọng số trung bình
```

```python
print(confusion_matrix(y_test, y_pred)) #Confusion Matrix ma trận nhầm lẫn hay ma trận lỗi
```
